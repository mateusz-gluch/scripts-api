```python
import requests
import io

from datetime import datetime, UTC
from loguru import logger
import pandas as pd
```

```python
URL = "https://dev-internal-api.elmodis.com/scripts/events-summary/data"
```

### Timestamp testing

No timestamp

```python
resp = requests.get(URL, params={"assets": "210"})
buf = io.BytesIO(resp.content)

pd.read_json(buf)

# correct representation of data
```

Set timestamp

```python
start = datetime(2024,4,2, tzinfo=UTC)
end = datetime(2024,4,3,second=1, tzinfo=UTC)
# ok
resp = requests.get(URL, params={"assets": "210", "ts":f"{int(start.timestamp())}:{int(end.timestamp())}"})
buf = io.BytesIO(resp.content)
pd.read_json(buf)
```

### Category testing

Single Category

```python
start = datetime(2024,4,3, tzinfo=UTC) # utc
end = datetime(2024,4,3,second=1, tzinfo=UTC)
tsStr = f"{int(start.timestamp())}:{int(end.timestamp())}"

resp = requests.get(URL, params={"assets": "210", "category": "MACHINE", "ts": tsStr})

buf = io.BytesIO(resp.content)
pd.read_json(buf)
```

Multiple Categories

```python
start = datetime(2024,4,3, tzinfo=UTC) # utc
end = datetime(2024,4,3,second=1, tzinfo=UTC)
tsStr = f"{int(start.timestamp())}:{int(end.timestamp())}"

resp = requests.get(URL, params={"assets": "210", "category": "MACHINE,DIAGNOSTICS", "ts": tsStr})

buf = io.BytesIO(resp.content)
pd.read_json(buf)
```

### Accept testing

Json

CSV

```python
resp = requests.get(URL, params={"assets": "210", "category":"MACHINE"}, headers={"accept": "text/csv"})
buf = io.BytesIO(resp.content)
pd.read_csv(buf)
```
